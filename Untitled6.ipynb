```python
import cv2
import matplotlib.pyplot as plt

face_cascade=cv2.CascadeClassifier("/content/face_cascade classifier.xml")
face_cascade
```

```python

eye_cascade=cv2.CascadeClassifier("/content/eye_cascade classifier.xml")
eye_cascade
```

```python
person= cv2.imread("/content/pexels-photo-220453.jpeg")
```

```python
plt.imshow(person)
```

```python
gray=cv2.cvtColor(person,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray)
```

```python
faces = face_cascade.detectMultiScale(gray, 1.1, 4)
```

```python
faces
```

```python
eyes =eye_cascade.detectMultiScale(gray, 1.1, 4)
eyes
```

```python
color = cv2.cvtColor(person, cv2.COLOR_BGR2RGB)
```

```python
for(x,y,w,h) in faces:
  cv2.rectangle(person, (x,y), (x+w,y+h), (255,0,0))
  plt.imshow(color)
  cv2.waitKey()
```
